```python
from hypothetical.nonparametric import WilcoxonTest
def apply_police(df):
    experiment_name = df["experiment"].iloc[0]
    df = df.copy()
    df = df[(df["epoch"] > 5) & (df["holdout_matthews_corrcoef"] > 0.05)]
    df = df.sort_values("run")

    if "politica_1" in experiment_name:
        r = df.sort_values(["test_overall_acc", "num_neurons"], ascending=[False, True])
    elif "politica_2" in experiment_name:
        r = df.sort_values(["holdout_overall_acc", "num_neurons"], ascending=[False, True])
    elif "politica_3" in experiment_name:
        r = df.sort_values(["validation_overall_acc", "num_neurons"], ascending=[False, True])
    elif "politica_4" in experiment_name:
        r = df.sort_values(["validation_overall_acc", "num_neurons"], ascending=[False, True])
    else:
        raise RuntimeError(f"No policy found in experiment name {experiment_name}.")

    r = r.loc[r.groupby(["experiment", "dataset", "run"])["test_overall_acc"].idxmax()]
    # r.columns = [''.join(c) for c in r.columns.values]
    # r.columns = [c.replace("max", "") for c in r.columns]

    return r

tmp_df = df.copy()
tmp_df = tmp_df[~tmp_df["dataset"].isin(["libras_move", "ilpd"])]
# tmp_df = tmp_df[
# 	(df["experiment"] == "exp0087_politica_1_oracle_1m1l") &
# 	(df["dataset"] == "Australian")
# ]
# print(tmp_df)
# tmp_df = apply_police(tmp_df)
# tmp_df.sort_values("run")
tmp_df.sort_values(["run"])
tmp_df = tmp_df[tmp_df["experiment"].str.contains("politica_1")]
for dataset in tmp_df["dataset"].unique():
    df_dataset = tmp_df[tmp_df["dataset"] == dataset]
    for exp1 in tmp_df["experiment"].unique():
        d1 = df_dataset[df_dataset["experiment"] == exp1]
        d1 = apply_police(d1)
        for exp2 in tmp_df["experiment"].unique():
            if exp1 != exp2:
                d2 = df_dataset[df_dataset["experiment"] == exp2]
                d2 = apply_police(d2)
                print(f'{dataset} - {exp1} - {exp2}')
                x1 = d1["test_overall_acc"]
                x2 = d2["test_overall_acc"]
                print(f"{x1.shape}, {x2.shape}")
                test = WilcoxonTest(x1, x2)
                print(f'({d1["test_overall_acc"].mean()}) - ({d2["test_overall_acc"].mean()})  {test.test_summary}')
                # print(f'{dataset} - {exp1} ({d1["test_overall_acc"].mean()}) - {exp2} ({d2["test_overall_acc"].mean()})  {test.test_summary}')
```
